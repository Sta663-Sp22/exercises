```python
import numpy as np
import pandas as pd

import pyarrow as pa

import os
```

```python
import pyarrow.dataset as ds
```

```python
dataset = ds.dataset("s3://ursa-labs-taxi-data/2019/")
```

```python
dataset.files
```

```python
dataset.schema
```

```python
import pyarrow.compute as pc
```

```python
col_sum = 0
count = 0
i = 0
for batch in dataset.to_batches(columns=["fare_amount"]):
    col_sum += pc.sum(batch.column("fare_amount")).as_py()
    count += batch.num_rows
    print("Batch ", i)
    i += 1

col_sum / count
```

```python
dataset.to_table(filter = ds.field("fare_amount") > 100).to_pandas()
```
