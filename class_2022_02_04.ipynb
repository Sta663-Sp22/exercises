```python
import numpy as np
import pandas as pd

df = pd.read_csv("https://sta663-sp22.github.io/slides/data/us_rent.csv")
```

```python
( df
  .pivot(index="name", columns="variable", values="estimate")
  .assign(
      ratio = lambda d: d.income / d.rent
  )
  .reset_index()
  .sort_values("ratio", ascending = True)
)
```
